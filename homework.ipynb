Find a real dataset using singly linked list, doubly linked list, and Circular Linked List.



*   Singly Linked List





```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None
class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        cur = self.head
        while cur.next:
            cur = cur.next
        cur.next = new_node

    def delete(self, key, col_name):
        cur = self.head
        if cur and cur.data[col_name] == key:
            self.head = cur.next
            return
        prev = None
        while cur and cur.data[col_name] != key:
            prev = cur
            cur = cur.next
        if cur:
            prev.next = cur.next

    def print_list(self):
        cur = self.head
        print("=== Singly Linked List ===")
        while cur:
            print(cur.data)
            cur = cur.next
        print()
```



* Doubly Linked List

```python
class DNode:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None


class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = DNode(data)
        if not self.head:
            self.head = new_node
            return
        cur = self.head
        while cur.next:
            cur = cur.next
        cur.next = new_node
        new_node.prev = cur

    def print_forward(self):
        print("=== Doubly Linked List (Forward) ===")
        cur = self.head
        last = None
        while cur:
            print(cur.data)
            last = cur
            cur = cur.next
        print()
        return last

    def print_backward(self, last):
        print("=== Doubly Linked List (Backward) ===")
        cur = last
        while cur:
            print(cur.data)
            cur = cur.prev
        print()
```



*   Circuly Linked List



```python
class CNode:
    def __init__(self, data):
        self.data = data
        self.next = None


class CircularLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = CNode(data)
        if not self.head:
            self.head = new_node
            new_node.next = self.head
            return
        cur = self.head
        while cur.next != self.head:
            cur = cur.next
        cur.next = new_node
        new_node.next = self.head

    def print_list(self):
        print("=== Circular Linked List ===")
        if not self.head:
            print("List is empty")
            return
        cur = self.head
        while True:
            print(cur.data)
            cur = cur.next
            if cur == self.head:
                break
        print("(Back to head)\n")
```

```python
# Convert each row to a dictionary
records = data.to_dict(orient="records")

# Create and fill the Singly Linked List
sll = SinglyLinkedList()
for record in records:
    sll.append(record)
sll.print_list()

# Example delete
if "Student_ID" in cols:
    delete_key = data.iloc[0]["Student_ID"]
    sll.delete(delete_key, "Student_ID")
    print(f"After deleting Student_ID = {delete_key}:")
    sll.print_list()


# Create and fill the Doubly Linked List
dll = DoublyLinkedList()
for record in records:
    dll.append(record)
last = dll.print_forward()
dll.print_backward(last)

# Create and fill the Circular Linked List
cll = CircularLinkedList()
for record in records:
    cll.append(record)
cll.print_list()

```
